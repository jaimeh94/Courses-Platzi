```python
import torch
import torch.nn as nn # modulo para redes neuronales
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Red Lineal
linear = nn.Linear(in_features=4096, out_features=10)
# Red convolucional
conv = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, stride=1, padding=1)
#Red relu
relu = nn.ReLU()
```

```python
x = torch.randn(4096)
y = linear(x)
y.size()
```

```python
x = torch.randn(1,3,7,7)
y = conv(x)
z = relu(x)
```
